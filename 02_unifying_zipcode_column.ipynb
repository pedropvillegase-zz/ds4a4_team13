````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.options.display.max_rows = None
````

### Cleaning ```zipcode``` real_state

````python
real_estate = pd.read_csv("datasets/real_estate.csv")
real_estate.head()
````

````python
real_estate['zipcode'] = real_estate['zipcode'].astype('str')

real_estate['zipcode_clean'] = real_estate['zipcode']
#real_estate[real_estate['len'] < 5]
real_estate['zipcode'][real_estate['zipcode'].isin(['nan', '1m', ' '])] = ''
real_estate['len'] = real_estate['zipcode'].apply( lambda x : len(str(x)))

real_estate['zipcode_clean'][real_estate['len'] == 1] = real_estate['zipcode'][real_estate['len'] == 1].apply( lambda x : '0000'+str(x))
real_estate['zipcode_clean'][real_estate['len'] == 2] = real_estate['zipcode'][real_estate['len'] == 2].apply( lambda x : '000'+str(x))
real_estate['zipcode_clean'][real_estate['len'] == 3] = real_estate['zipcode'][real_estate['len'] == 3].apply( lambda x : '00'+str(x))
real_estate['zipcode_clean'][real_estate['len'] == 4] = real_estate['zipcode'][real_estate['len'] == 4].apply( lambda x : '0'+str(x))
real_estate['zipcode_clean'] = list(real_estate['zipcode_clean'].str.slice(0,5))

real_estate.drop(['len'], axis = 1, inplace = True)
real_estate[['zipcode', 'zipcode_clean']][real_estate['zipcode'] != real_estate['zipcode_clean']]
````

### Cleaning ```zipcode``` listings

````python
listings = pd.read_csv("datasets/listings.csv")
listings.head()
````

````python
listings['zipcode'] = listings['zipcode'].astype('str')

listings['zipcode_clean'] = listings['zipcode']
#listings[listings['len'] < 5]
listings['zipcode'][listings['zipcode'].isin(['nan', '1m', ' '])] = ''
listings['len'] = listings['zipcode'].apply( lambda x : len(str(x)))

listings['zipcode_clean'][listings['len'] == 1] = listings['zipcode'][listings['len'] == 1].apply( lambda x : '0000'+str(x))
listings['zipcode_clean'][listings['len'] == 2] = listings['zipcode'][listings['len'] == 2].apply( lambda x : '000'+str(x))
listings['zipcode_clean'][listings['len'] == 3] = listings['zipcode'][listings['len'] == 3].apply( lambda x : '00'+str(x))
listings['zipcode_clean'][listings['len'] == 4] = listings['zipcode'][listings['len'] == 4].apply( lambda x : '0'+str(x))
listings['zipcode_clean'] = list(listings['zipcode_clean'].str.slice(0,5))

listings.drop(['len'], axis = 1, inplace = True)
listings[['zipcode', 'zipcode_clean']][listings['zipcode'] != listings['zipcode_clean']].dropna()
````

### Cleaning ```zipcode``` demographics

````python

demographics = pd.read_csv("datasets/demographics.csv")
demographics.head()
````

````python
demographics['zipcode'] = demographics['zipcode'].astype('str')

demographics['zipcode_clean'] = demographics['zipcode']
#demographics[demographics['len'] < 5]
demographics['zipcode'][demographics['zipcode'].isin(['nan', '1m', ' '])] = ''
demographics['len'] = demographics['zipcode'].apply( lambda x : len(str(x)))

demographics['zipcode_clean'][demographics['len'] == 1] = demographics['zipcode'][demographics['len'] == 1].apply( lambda x : '0000'+str(x))
demographics['zipcode_clean'][demographics['len'] == 2] = demographics['zipcode'][demographics['len'] == 2].apply( lambda x : '000'+str(x))
demographics['zipcode_clean'][demographics['len'] == 3] = demographics['zipcode'][demographics['len'] == 3].apply( lambda x : '00'+str(x))
demographics['zipcode_clean'][demographics['len'] == 4] = demographics['zipcode'][demographics['len'] == 4].apply( lambda x : '0'+str(x))
demographics['zipcode_clean'] = list(demographics['zipcode_clean'].str.slice(0,5))

demographics.drop(['len'], axis = 1, inplace = True)
demographics[['zipcode', 'zipcode_clean']][demographics['zipcode'] != demographics['zipcode_clean']].dropna()
````

### ```Zipcode``` joins listings and demographics

````python
print("listings shape: " +  str(listings.shape))
print("demographics shape: " +  str(demographics.shape))
print("left join between listings and demographics:" +
    str(pd.DataFrame(
    listings['zipcode_clean'])
 .merge(
     pd.DataFrame(demographics['zipcode_clean']),
     how = 'left',
     on = 'zipcode_clean' ).shape)
)
print("inner join between listings and demographics:" +
    str(pd.DataFrame(
    listings['zipcode_clean'])
 .merge(
     pd.DataFrame(demographics['zipcode_clean']),
     how = 'inner',
     on = 'zipcode_clean' ).shape)
)

print("right join between listings and demographics:" +
    str(pd.DataFrame(
    listings['zipcode_clean'])
 .merge(
     pd.DataFrame(demographics['zipcode_clean']),
     how = 'right',
     on = 'zipcode_clean' ).shape)
)
````
