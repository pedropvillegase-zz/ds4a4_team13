````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.options.display.max_rows = None
````

### Cleaning ```zipcode``` real_state

````python
real_estate = pd.read_csv("datasets/real_estate.csv")
real_estate.head()
````

````python
real_estate['zipcode'] = real_estate['zipcode'].astype('str')

#real_estate[real_estate['len'] < 5]
real_estate['zipcode_clean'] = real_estate['zipcode']
real_estate['zipcode_clean'][real_estate['zipcode_clean'].isin(['nan', '1m', ' '])] = ''

real_estate['zipcode_clean'] = real_estate['zipcode_clean'].apply( lambda x : x.zfill(5))
real_estate['zipcode_clean'] = list(real_estate['zipcode_clean'].str.slice(0,5))

(real_estate[
    ['zipcode', 'zipcode_clean']
][real_estate['zipcode'] != real_estate['zipcode_clean']]
  .drop_duplicates()
 .sort_values(by = 'zipcode'))
````

### Cleaning ```zipcode``` listings

````python
listings = pd.read_csv("datasets/listings.csv")
listings.head()
````

````python
listings['zipcode'] = listings['zipcode'].astype('str')

#listings[listings['len'] < 5]
listings['zipcode_clean'] = listings['zipcode']
listings['zipcode_clean'][listings['zipcode_clean'].isin(['nan', '1m', ' '])] = ''

listings['zipcode_clean'] = listings['zipcode_clean'].apply( lambda x : x.zfill(5))
listings['zipcode_clean'] = list(listings['zipcode_clean'].str.slice(0,5))

(listings[
    ['zipcode', 'zipcode_clean']
][listings['zipcode'] != listings['zipcode_clean']]
  .drop_duplicates()
 .sort_values(by = 'zipcode'))
````

### Cleaning ```zipcode``` demographics

````python
demographics = pd.read_csv("datasets/demographics.csv")
demographics.head()
````

````python
demographics['zipcode'] = demographics['zipcode'].astype('str')

#demographics[demographics['len'] < 5]
demographics['zipcode_clean'] = demographics['zipcode']
demographics['zipcode_clean'][demographics['zipcode_clean'].isin(['nan', '1m', ' '])] = ''

demographics['zipcode_clean'] = demographics['zipcode_clean'].apply( lambda x : x.zfill(5))
demographics['zipcode_clean'] = list(demographics['zipcode_clean'].str.slice(0,5))

(demographics[
    ['zipcode', 'zipcode_clean']
][demographics['zipcode'] != demographics['zipcode_clean']]
  .drop_duplicates()
 .sort_values(by = 'zipcode'))
````

### ```Zipcode``` joins listings and demographics

````python
print("listings shape: " +  str(listings.shape))
print("demographics shape: " +  str(demographics.shape))
print("left join between listings and demographics:" +
    str(pd.DataFrame(
    listings['zipcode_clean'])
 .merge(
     pd.DataFrame(demographics['zipcode_clean']),
     how = 'left',
     on = 'zipcode_clean' ).shape)
)
print("inner join between listings and demographics:" +
    str(pd.DataFrame(
    listings['zipcode_clean'])
 .merge(
     pd.DataFrame(demographics['zipcode_clean']),
     how = 'inner',
     on = 'zipcode_clean' ).shape)
)

print("right join between listings and demographics:" +
    str(pd.DataFrame(
    listings['zipcode_clean'])
 .merge(
     pd.DataFrame(demographics['zipcode_clean']),
     how = 'right',
     on = 'zipcode_clean' ).shape)
)
````

### ```Zipcode``` joins listings and real_estate

````python
print("listings shape: " +  str(listings.shape))
print("real_estate shape: " +  str(real_estate.shape))
print("left join between listings and real_estate:" +
    str(pd.DataFrame(
    listings['zipcode_clean'])
 .merge(
     pd.DataFrame(real_estate['zipcode_clean']),
     how = 'left',
     on = 'zipcode_clean' ).shape)
)
print("inner join between listings and real_estate:" +
    str(pd.DataFrame(
    listings['zipcode_clean'])
 .merge(
     pd.DataFrame(real_estate['zipcode_clean']),
     how = 'inner',
     on = 'zipcode_clean' ).shape)
)

print("right join between listings and real_estate:" +
    str(pd.DataFrame(
    listings['zipcode_clean'])
 .merge(
     pd.DataFrame(real_estate['zipcode_clean']),
     how = 'right',
     on = 'zipcode_clean' ).shape)
)
````
